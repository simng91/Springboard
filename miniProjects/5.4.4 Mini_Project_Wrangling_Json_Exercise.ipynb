## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
data
```

```python
# use normalization to create tables from nested element
pd.json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
pd.json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
json_df = pd.read_json('data/world_bank_projects.json')
json_df.countryname.value_counts().head(10)
```

```python
json_string = json.load((open('data/world_bank_projects.json')))
code_names = pd.json_normalize(json_string, 'mjtheme_namecode')
top10_codes = code_names.code.value_counts()
df1 = top10_codes[0:10].index.to_frame(name="code")
d = top10_codes[0:10].to_dict()
df2 = code_names[(code_names.code.isin(d)) & (code_names.name != "")].drop_duplicates()
top10 = pd.merge(df1, df2, on="code", how="outer")
top10
```

```python
normalized_data = pd.json_normalize(json_string, 'mjtheme_namecode', ['_id', 'project_name'])

# pd.json_normalize(json_string, 'mjtheme_namecode', ['sector', 'projectfinancialtype', 'supplementprojectflg'])
#        'supplementprojectflg', 'projectfinancialtype', 'prodline', 'mjtheme', 'idacommamt', 'impagency', 'project_name', 'mjthemecode',
#        'closingdate', 'totalcommamt', 'id', 'mjsector_namecode', 'docty',
#        'sector1', 'lendinginstr', 'countrycode', 'sector2', 'totalamt',
#        'mjtheme_namecode', 'boardapprovaldate', 'countryshortname', 'sector4',
#        'prodlinetext', 'productlinetype', 'regionname', 'status',
#        'country_namecode', 'envassesmentcategorycode', 'project_abstract',
#        'approvalfy', 'projectdocs', 'lendprojectcost', 'lendinginstrtype',
#        'theme1', 'grantamt', 'themecode', 'borrower', 'sectorcode', 'sector3',
#        'majorsector_percent', 'board_approval_month', 'theme_namecode',
#        'countryname', 'url', 'source', 'projectstatusdisplay', 'ibrdcommamt',
#        'sector_namecode', '_id'])   --> this gave errors
```

```python
table = code_names[code_names.name != ""].drop_duplicates()
#table.code = pd.to_numeric(table.code)
table = table.set_index('code').sort_index().to_dict()
table['name']
```

```python
normalized_data['name'] = normalized_data['code'].apply(lambda x: table['name'][x])
# proof that there is no missing name in any row
normalized_data[normalized_data['name'] == ""]
```

```python
normalized_data = normalized_data.sort_values(by=['project_name'])
normalized_data2 = pd.json_normalize(json_string)

final_df = pd.merge(normalized_data, normalized_data2, on="project_name", how="outer")
final_df
```
